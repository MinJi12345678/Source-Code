```python
import pandas as pd

DF = pd.read_csv('/content/drive/MyDrive/5조/03. 데이터/날씨 데이터/23.03.21~24.03.20.csv', encoding='cp949')
```

```python
DF['강수량(mm)'].fillna(0, inplace = True)
DF['적설(cm)'].fillna(0, inplace = True)
DF.head()
```

```python
DF['일시'] = pd.to_datetime(DF['일시'])

DF['주'] = DF['일시'].dt.week
```

```python
DF.info()
```

```python
DF.head(500)
```

```python
def calculate_wind_chill(temperature, wind_speed):
    wind_chill = 13.12 + 0.6215 * temperature - 11.37 * (wind_speed ** 0.16) + 0.3965 * temperature * (wind_speed ** 0.16)
    return wind_chill
```

```python
DF['체감온도'] = calculate_wind_chill(DF['기온(°C)'], DF['풍속(m/s)'])

DF.head()
```

```python
DF = DF.groupby(pd.Grouper(key='일시', freq='W-WED', label='left', closed='left')).agg({'기온(°C)': ['max', 'min', 'mean'],
                                                                                      '강수량(mm)' : ['sum'],
                                                                                      '습도(%)':['mean'],
                                                                                      '적설(cm)':['sum'],
                                                                                      '체감온도':['mean']}).round(2)
```

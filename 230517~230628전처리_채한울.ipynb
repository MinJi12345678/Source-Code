# 1. 2023년 6월 14일 패션 데이터셋

```python
import pandas as pd

DF1 = pd.read_csv('/content/drive/MyDrive/5조/03. 데이터/패션 데이터/230614.csv')
```

```python
DF1.info()
```

```python
DF1.head(10)
```

```python
DF1.isnull().sum(axis = 0)
```

```python
columns_to_drop = ['category_large_desc', 'use','print_pattern','characteristic', 'detail',
                   'age', 'weight', 'season', 'sensibility']

DF1.drop(columns = columns_to_drop, inplace = True)
```

```python
DF1['item'].fillna(DF1['category_small_desc'], inplace = True)
```

```python
DF1.dropna(inplace = True)
```

```python
DF1 = DF1[DF1['recent_sale_count'] >= 1]
```

```python
DF1.info()
```

```python
DF1.insert(0, 'Time', '2023_06_14')

DF1.head()
```

```python
DF1.to_csv('/content/drive/MyDrive/5조/03. 데이터/1차 전처리 패션 데이터/230614.csv' , index = False)
```

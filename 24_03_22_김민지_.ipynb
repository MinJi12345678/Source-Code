3) 데이터 복사

4) 불필요 데이터 삭제

```python
df1 = df1.drop(df1.columns[[0, 5, 7, 8, 9, 10]], axis=1)
df1.head(5)
```

```python
df1['GOODS'].fillna(df1['SCLAS'], inplace = True)
```

```python
df1.isnull().sum(axis = 0)
```

```python
df1.dropna(axis = 0, inplace = True)
```

```python
df1.isnull().sum(axis = 0)
```

```python
df1.info()
```

```python
df1 = df1[df1['LATST'] >= 1]
```

```python
df1.insert(0, 'Time', '2023_08_23')

df1.head(3)
```

```python
names = {'MLSFC':'category_middle_desc', 'SCLAS':'category_small_desc', 'GOODS':'item',
                    'COLOR':'color', 'QLFMT':'material', 'SLPR':'sale_price', 'LATST':'recent_sale_count'}

df1.rename(columns = names, inplace = True)
df1.head(3)
```
